```python
#Componentes de cores da imagem, apresentar as três componentes de cores da imagem

img_2 = cv2.imread('/content/drive/MyDrive/PIM_imagens/torre_eiffel.webp')


# Deixa apenas o canal Azul
img_only_blue = copy.copy(img_2)

#img_only_blue[:,:,0] = 0 
img_only_blue[:,:,1] = 0
img_only_blue[:,:,2] = 0


# Deixa apenas o canal VERDE
img_only_green = copy.copy(img_2)

img_only_green[:,:,0] = 0
#img_only_green[:,:,1] = 0
img_only_green[:,:,2] = 0

# Deixa apenas o canal VERMELHO
img_only_red = copy.copy(img_2)

img_only_red[:,:,0] = 0
img_only_red[:,:,1] = 0
#img_only_red[:,:,2] = 0

# Exibe as imagens
cv2_imshow(img_2 )
cv2_imshow(img_only_blue)
cv2_imshow(img_only_green) 
cv2_imshow(img_only_red)
```
